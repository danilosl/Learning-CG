> **Projeto 2**: Realizar as diferentes transformações geométricas em 2D:
*   Translação
*   Escala
*   Reflexão
*   Rotação
*   Cisalhamento

```python
import numpy as np
import matplotlib.pyplot as plt
```

###**Defindo os pontos**

```python
x_axis=np.array([[0, 1, 2, 3],
                 [0, 0, 0, 0],
                 [0, 1, 2, 3],
                 [3, 3, 3, 3]])

y_axis=np.array([[3, 3, 3, 3],
                 [0, 1, 2, 3],
                 [0, 0, 0, 0],
                 [0, 1, 2, 3]])

color_ = ['maroon',
          'g',
          'r',
          'c']

for i in range(4):
  plt.plot(x_axis[i][:], y_axis[i][:],color=color_[i])
plt.title('Imagem Original', fontsize=14, weight='bold')
plt.grid()
plt.show()
```

###**Translação**

```python
for i in range(4):
  plt.plot(x_axis[i][:]+10, y_axis[i][:]+10, color=color_[i])
plt.title('Translação em 10 unidades da Imagem Original', fontsize=14, weight='bold')
plt.grid()
plt.show()
```

###**Escala**

```python
for i in range(4):
  plt.plot(x_axis[i][:]*1.5, y_axis[i][:]*1.5, color=color_[i])
plt.title('Aumentando a Imagem Original em 50%', fontsize=14, weight='bold')
plt.grid()
plt.show()
```

###**Reflexão**

```python
for i in range(4):
  plt.plot(x_axis[i][:]*-1, y_axis[i][:]*-1, color=color_[i])
plt.title('Reflexão da Imagem Original', fontsize=14, weight='bold')
plt.grid()
plt.show()
```

###**Rotação**
$$ x '= x \cos(θ) - y\sin(\theta) \\ y' = y\cos(\theta) + x\sin(\theta)$$

```python
for i in range(4):
  plt.plot(x_axis[i][:]*np.cos(np.pi/4) - y_axis[i][:]*np.sin(np.pi/4), y_axis[i][:]*np.cos(np.pi/4) + x_axis[i][:]*np.sin(np.pi/4), color=color_[i])
plt.title('Rotação em 45° da Imagem Original', fontsize=14, weight='bold')
plt.grid()
plt.show()
```

###**Cisalhamento**
$$ x '= x + k_{x}y \\ y' = y + k_{y}x$$

```python
# definindo os valores de Kx e Ky
Kx, Ky = 1.5, 0
for i in range(4):
  plt.plot(x_axis[i][:] + Kx * y_axis[i][:], y_axis[i][:] + Ky * x_axis[i][:], color=color_[i])
plt.title('Cisalhamento da Imagem Original', fontsize=14, weight='bold')
plt.grid()
plt.show()
```
